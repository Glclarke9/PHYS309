```python
# Part A
#proton speed
u_prime = 0.99 * constants.c
#array of v values
v = np.arange(1,1000,1)
#calculating proton speed to frank
Fvel = (u_prime+v)/(1+((v*u_prime)/constants.c**2))
#plotting
plt.plot(v,Fvel)
plt.xlabel('Spacecraft speed')
plt.ylabel('Velocity measured by Frank')
plt.show()
#Part B
#calculating gamma
gamma = (1-(v**2/constants.c**2))**(1/2)
#calculating new u_prime with reference to the conical shape
uy_prime = u_prime*math.sin(.261799)
#calcularting new velocity
u_y = (uy_prime*gamma)/(1+(v/constants.c**2)*(u_prime*math.cos(.261799)))
#plotting
plt.plot(v,u_y)
plt.xlabel('Spacecraft Speed')
plt.ylabel('YVel measured by Frank')
plt.show()
```
